```python
#Média e Mediana de salários

salarios = [45000, 15000, 10000, 10000, 5700, 5000, 5000, 3700, 3000, 2000, 2000, 2000, 2000]

#Média
n = len(salarios)
soma = 0
for salario in salarios:
    soma += salario
media = soma / n

#Variância
soma_diferencas = 0
for salario in salarios:
    soma_diferencas += (salario - media) ** 2
variancia = soma_diferencas / n

#Desvio Padrão
desvio_padrao = variancia ** 0.5

print(f"Média: {media:.2f}")
print(f"Variância: {variancia:.2f}")
print(f"Desvio Padrão: {desvio_padrao:.2f}")

#Mediana
salarios_ordenados = sorted(salarios)
meio = n // 2
if n % 2 == 0: 
    mediana = (salarios_ordenados[meio - 1] + salarios_ordenados[meio]) / 2
else:  # número ímpar
    mediana = salarios_ordenados[meio]

#Moda
contagens = {}
for s in salarios:
    if s in contagens:
        contagens[s] += 1
    else:
        contagens[s] = 1

#Valor com maior frequência
moda = None
maior_freq = 0
for valor, freq in contagens.items():
    if freq > maior_freq:
        maior_freq = freq
        moda = valor

print(f"Mediana: {mediana:.2f}")
print(f"Moda: {moda} (frequência: {maior_freq})")
```

# Célula 2 — Cálculo manual de média, variância, mediana e moda (salários)

O que faz:
- Define uma lista `salarios` e calcula média, variância, desvio padrão, mediana e moda por algoritmos manuais (laços), demonstrando fórmulas passo-a-passo.

Entradas/Saídas: entrada = lista `salarios`; saída = valores impressos de média, variância, desvio, mediana e moda.

Observação: objetivo didático para entender como as estatísticas são calculadas sem usar funções prontas.

```python
#Gerar numeros aleatórios
import random, math
import matplotlib.pyplot as plt

N = 10000
dados = [random.randint(0, 10) for _ in range(N)]

# mín e máx
_min = dados[0]; _max = dados[0]
for x in dados:
    if x < _min: _min = x
    if x > _max: _max = x

# média
soma = 0
for x in dados: soma += x
media = soma / N

# variância e desvio
soma_dif = 0.0
for x in dados: soma_dif += (x - media) ** 2
variancia = soma_dif / N
desvio = math.sqrt(variancia)

# mediana
dados_ord = sorted(dados)
if N % 2 == 1:
    mediana = dados_ord[N//2]
else:
    mediana = (dados_ord[N//2 - 1] + dados_ord[N//2]) / 2

# moda
freq = {}
for x in dados:
    freq[x] = freq.get(x, 0) + 1
maior = max(freq.values())
modas = [k for k,v in freq.items() if v == maior]

# percentuais
def pct_dentro(ds, mean, sigma, k):
    lo = mean - k*sigma
    hi = mean + k*sigma
    cnt = sum(1 for v in ds if lo <= v <= hi)
    return cnt / len(ds) * 100, lo, hi

p1, lo1, hi1 = pct_dentro(dados, media, desvio, 1)
p2, lo2, hi2 = pct_dentro(dados, media, desvio, 2)
p3, lo3, hi3 = pct_dentro(dados, media, desvio, 3)

print("N:", N)
print("min:", _min, "max:", _max, "média:", media)
print("mediana:", mediana)
print("moda(s):", modas, "freq:", maior)
print("variância:", variancia, "desvio:", desvio)
print("±1σ:", f"{p1:.2f}% — [{lo1:.3f}, {hi1:.3f}]")
print("±2σ:", f"{p2:.2f}% — [{lo2:.3f}, {hi2:.3f}]")
print("±3σ:", f"{p3:.2f}% — [{lo3:.3f}, {hi3:.3f}]")

#Histograma
plt.figure(figsize=(8,4))
plt.hist(dados, bins=range(0,12), edgecolor='black')  # bins para inteiros 0..10
plt.title("Histograma dos 10.000 inteiros aleatórios (0 a 10)")
plt.xlabel("Valor")
plt.ylabel("Frequência")
plt.xticks(range(0,11))
plt.show()
```

# Célula 3 — Geração de números aleatórios, estatísticas e histograma

O que faz:
- Gera N números inteiros aleatórios (0..10), calcula mínimo, máximo, média, variância, desvio, mediana, moda e percentuais dentro de k sigmas.
- Plota um histograma dos valores gerados.

Observação: demonstra propriedades da distribuição empírica e a regra empírica (±1σ, ±2σ, ±3σ).

```python
#Alturas de Homens
import numpy as np
import matplotlib.pyplot as plt

#Dados simulados
mu_h, sigma_h = 175, 6   # média e desvio padrão
dados_homens = np.random.normal(mu_h, sigma_h, 10000)

media_h = np.mean(dados_homens)
variancia_h = np.var(dados_homens)
desvio_h = np.std(dados_homens)

print("Altura (Homens)")
print(f"Média: {media_h:.2f} cm")
print(f"Variância: {variancia_h:.2f}")
print(f"Desvio padrão: {desvio_h:.2f}\n")

#Histograma
plt.hist(dados_homens, bins=30, color='lightblue', edgecolor='black', density=True)
plt.title("Distribuição de Altura - Homens")
plt.xlabel("Altura (cm)")
plt.ylabel("Frequência relativa")
plt.grid(True)
plt.show()

#Intervalos da regra empírica (68-95-99.7)
intervalo_68 = (mu_h - sigma_h, mu_h + sigma_h)
intervalo_95 = (mu_h - 2*sigma_h, mu_h + 2*sigma_h)
intervalo_997 = (mu_h - 3*sigma_h, mu_h + 3*sigma_h)

print("Intervalos empíricos (Homens):")
print(f"68% dos dados entre {intervalo_68[0]:.2f} e {intervalo_68[1]:.2f} cm")
print(f"95% dos dados entre {intervalo_95[0]:.2f} e {intervalo_95[1]:.2f} cm")
print(f"99.7% dos dados entre {intervalo_997[0]:.2f} e {intervalo_997[1]:.2f} cm")
```

# Célula 5 — Simulação de alturas (Mulheres)

O que faz:
- Gera uma amostra normal simulada de alturas femininas (média 162 cm, sd 5 cm), calcula estatísticas e plota histograma; exibe intervalos empíricos.

Observação: paralelo à análise de homens para comparação entre distribuições.

```python
#Propaganda e Lucro
import pandas as pd
import matplotlib.pyplot as plt

#DataFrame com os dados
dados = {
    "Custo com propaganda": [1000, 1150, 1500, 2000, 2200, 10000],
    "Lucro sobre cada item": [5500, 5400, 5600, 5900, 6000, 20000]
}

df = pd.DataFrame(dados)
print("=== Dados da Tabela A ===")
print(df)

# Gerar gráfico de dispersão
df.plot(
    kind="scatter",
    x="Custo com propaganda",
    y="Lucro sobre cada item",
    title="Relação entre Custo com Propaganda e Lucro",
    color="royalblue"
)
plt.grid(True)
plt.show()

# Calcular correlação
print("\n=== Correlações ===")
print("Correlação de Pearson:")
print(df.corr(method="pearson"))

print("\nCorrelação de Spearman:")
print(df.corr(method="spearman"))
```
